```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
absences = np.array([6,2,15,9,12,5,8])
grades = np.array([82,86,43,74,58,90,78])
```

```python
correlation, _ = pearsonr(absences, grades)
correlation
```

```python
plt.scatter(absences, grades)
plt.xlabel('Absences')
plt.ylabel('Grades')

m, b = np.polyfit(absences, grades, 1)
plt.plot(absences, m*absences + b, color="red")
```
