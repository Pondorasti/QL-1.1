```python
random_series = pd.Series(
    np.random.randn(1000),
    index=pd.date_range('1/1/2000', periods=1000)
)
random_series = random_series.cumsum()
random_series.plot()
```

```python
random_data_frame = pd.DataFrame(np.random.randn(1000, 6), index=random_series.index)
random_data_frame = random_data_frame.cumsum()
random_data_frame.plot()
plt.legend(loc="best")
```
