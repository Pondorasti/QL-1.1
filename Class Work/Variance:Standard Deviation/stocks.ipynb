```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_apple = pd.read_csv('AAPL_data.csv')
df_google = pd.read_csv('GOOGL_data.csv')

close_apple = df_apple['close']
close_google = df_google['close']
```

```python
plt.figure()

bins = np.linspace(0, 1000, 1000)

plt.hist(close_apple, bins, alpha = 0.5, label = 'Apple')
plt.hist(close_google, bins, alpha = 0.5, label = 'Google')

plt.show()
```

```python
print(f'Google STD: {close_google.std()}')
```
